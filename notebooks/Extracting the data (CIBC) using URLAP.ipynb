```python
from bs4 import BeautifulSoup
import requests
import os
import re
```

```python
url = "https://www.cibc.com/en/about-cibc/investor-relations/quarterly-results.html"
```

```python
req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')
```

```python
pdf_links = soup.find_all('a', href=True)
```

```python
pdfList = []
for link in pdf_links:
    if '.pdf' in link['href']: #and ('report' in link['href'] or 'ar-' in link['href']):
        pdfList.append('https://www.cibc.com' + link['href'])
```

```python
for i in range(len(pdfList)):
    response = requests.get(pdfList[i], stream=True)
    pdf_file_name = os.path.basename(pdfList[i])
#     print(pdf_file_name)
    if response.status_code == 200:
        year = re.findall(r'\d{4}', pdfList[i])
#         print(year)
        path = r'C:\Users\aanal\Documents\sem3\AI and ML Capstone Project\rag-chatbot-banks\data\CIBC\\' + year[0]
        if not os.path.exists(path):
            os.makedirs(path)
        filepath = os.path.join(path, pdf_file_name)
        with open(filepath, 'wb') as pdf_object:
            pdf_object.write(response.content)
            print('{}. {} was successfully saved!'.format(i, pdf_file_name))
    else:
        print(f'Uh oh! Could not download {pdf_file_name},')
        print(f'HTTP response status code: {response.status_code}')
```
